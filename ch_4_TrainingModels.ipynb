ùíö = Œ∏ùë• + Œ¶ \

# <div style="font-family: fantasy">Gradient Descent</div>

* The general idea is to tweak the parameters iteratively in order to minimize the cost function
* it easures the local gradient of the error function with regards to the parameter vector $\boldsymbol\theta$ and goes it goes in the direction ofdescending gradient.
* once the gradient is 0 it returns the $\boldsymbol\theta$ at that point
* we start with the random values of $\boldsymbol\theta$
* an important parameter in the gradient descent is the size of steps determined by the *learning rate* hyperparameter
* if learning rate is small the steps taken will be small and number of iterations will be large hece the running time wil be slow
* on the otherhand if the learning rate is to fast it might miss the minimum and again itertate toward decreasing gradient and thus diverges hence failing to find a good solution
* also there is a problem with random initial values the function might get stuck in local minima rather than global minima.
* also say if there is a plateu it will take long time addressing the descending gradient and thus will result into waste computations
* also to keep the curve of MSE regular we must standardize(or scale) our features in the data set 


### <div style="font-family: fantasy">batch gradient descent</div>
so we are intended in making the cost function minimum or the gradient 0\
thus we are interested in $\frac{\partial}{\partial \theta_j}\cdot \mathbf{MSE(\theta)}$\
and the equation giving it can be defined as


$\nabla_\theta MSE(\theta) = \frac{\partial}{\partial \theta}\cdot \mathbf{MSE(\theta)} = \frac{2}{m} \cdot \sum_{i=1}^{m}(\boldsymbol\theta^T \cdot X^i - y^i)\cdot x_j^i$ \
this is because,\
$\mathbf{MSE(\theta)} = \frac{1}{m}\cdot\sum_{i=1}^{m}(\Theta^T\cdot x^i - y^i)^2$\
differentiating it with respect to $\theta_j$ we get the equation (apply chain rule)


now instead of computing the gradient for all $\theta$ in the $\boldsymbol\theta$ we can compute them together 
$$gradient vector = \nabla_\theta MSE(\boldsymbol\theta) = 
\begin{pmatrix}
    \frac{\partial}{\partial \theta_0}\cdot \mathbf{MSE(\theta)} \\
    \frac{\partial}{\partial \theta_1}\cdot \mathbf{MSE(\theta)} \\
    \frac{\partial}{\partial \theta_2}\cdot \mathbf{MSE(\theta)} \\
                    \cdot \\
                    \cdot \\
                    \cdot \\
    \frac{\partial}{\partial \theta_n}\cdot \mathbf{MSE(\theta)} \\
\end{pmatrix}
 = \frac{2}{m}\cdot \mathbf{X^T} \cdot (\mathbf{X}\cdot \boldsymbol\theta - \boldsymbol y)
$$

Now once we get the gradient vector pointing uphill we need to start moving in th backward direction. Here comes therole of learning rate $\eta$ \
$\boldsymbol\theta^{(next step)} = \boldsymbol\theta - \eta \nabla_\theta MSE(\boldsymbol\theta)$ \
here   $\eta \nabla_\theta MSE(\boldsymbol\theta)$ serves as the $\Delta \theta$ for the eqution.\
hence,\
$\boldsymbol\theta^{(next step)} = \boldsymbol\theta - \Delta \boldsymbol\theta$


now lets look at the implementation of this algorithm

````python
eta = 0.1 #learning rate
n_iterations = 1000
m = 100
theta = np.random.rand(2,1)
for iter in range(n_iterations):
    print(theta)
    gradients = (2/m)*x_b.T.dot(x_b.dot(theta) - y)
    theta = theta - eta * gradients
````

the final theta with minimum gradient is there

````python
theta
````

note that $\boldsymbol\theta$ we found here is exactly equal to what we found with the normal equation \
also note the values may be different wrt to different $\eta$ \
now the problem lies with thenumber of iterations here we used 1000 bt there might be the case that this is not enough and also if we increase the iteration to a very large value we may end up wasting computations.\
thus the solution is to set a value $\epsilon$ called <em style="font-family: Tahoma, Geneva, Verdana, sans-serif;">**tolerance**</em> and interrupt the iterations when gradient vector becomes tiny - that is, when its norm becomes smaller than $\epsilon$. Because when this happens it means that gradient descent has almost reached minimum

### <div style="font-family: fantasy">Stocastic Gradient Descent </div>
* The benefit of stocastic gradient descent is it just randomly picks up an instance from the training set intead of taking the complete batch thus making this algoritm much faster since it has little data to manipulate at every iteration.
* it is suitable for working with large datasets since only ine isntance needs to be in teh memory at a time
* but due to stocastic nature this is much less regular than batch GD.
* instead of gently decreaing till minimum the cost function will bounce up and down decreasing only on average. 
* Over the time even after reaching the minimum it wont stop it will continue to bounce around and never settle down.
* once the algo stops parameter values are good but not optimal
* This also have the benefit that when cost function is irregular this propetry can help the algorithm to jump out of local minima so stocastic GD has a better chance of reacing the global minima.
* the randomness is good to escape from local optima but bad because algorithm will never settle at the minimum.
* the solution to this is to gradually reduce the learning rate.
* the steps start out large then get smaller and smaller allowing the algortihm to settle at the global minimum
* This process is called _simulated anhealing_
* the function that determines the learning rate at each schedule is called the _learning schedule_
* If learning rate is is reduced too slow you may jump around the minimum for a long time and end up with a suboptimal solution
* If it reduced too quickly we may get stuck in local minimum or even end up frozen halfway to the minimum

````python
#gradient descent 
import matplotlib as mpl
n_epocs = 50
t0,t1 = 5,50
def learning_schedule(t):
    return t0/(t+t1)
# random value initialization of theta
theta = np.random.randn(2,1)
for epoch in range(n_epocs):
    for i in range(m):
        random_index = np.random.randint(m)# selecting random instance from the training set
        # for plotting graph
        if epoch == 0 and i < 20:
            y_predict = x_inp_sim.dot(theta)
            color = mpl.colors.rgb2hex(plt.cm.OrRd(i / 20 + 0.15)) # type: ignore
            plt.plot(x_inp, y_predict, color=color)

        xi = x_b[random_index:random_index+1]
        yi = y[random_index:random_index+1]

        # setting the gradient
        gradients = (2)*xi.T.dot(xi.dot(theta) - yi) 
        # we do not use m because we are not using the total batch 
        #rather we iterate over every single instance
        eta = learning_schedule(epoch*m + 1)
        # setting up theta
        theta = theta - eta*gradients
plt.plot(x,y,'b.')
theta
````

now we can see the stocastic gd computer the correct values in 50 iterations while batch gd took 1000

but there can be another way i.e. looking at the learning curves


**learning curves**:
* these are plots of the model's performance on training and validation set as a function of the training set size(or the training iteration). (from performance we mean the errors )
* to generate these plots we train the model several times on different sized subsets of the training set

````python
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
def plot_learning_curve(model,X,y):
    X_train,X_val,y_train,y_val = train_test_split(X,y,train_size=0.8)
    train_errors,val_errors = [],[]
    for m in range(1,len(X_train)):
        model.fit(X_train[:m],y_train[:m])
        y_train_predict = model.predict(X_train[:m])
        y_val_predict = model.predict(X_val)
        train_errors.append(mean_squared_error(y_train[:m],y_train_predict))
        val_errors.append(mean_squared_error(y_val,y_val_predict))
    plt.plot(np.sqrt(train_errors),'r-',label='train')
    plt.plot(np.sqrt(val_errors),'-+',label='val',color='pink')
    plt.ylabel('RMSE')
    plt.ylim(0,3)
    plt.xlabel('training set size')
    plt.legend(loc='upper right')
````

````python
lin_reg_test = LinearRegression()
plot_learning_curve(lin_reg_test,X,y)
````

now we can notice that the training error is initially 0 which is obvious but increases as new data instances get added to the dataset.\
Similarly the validation error is initially high which then reduces to nealy equal to training set but still quite high.\
this proves that the model is underfitting

# <div style="font-family: fantasy;">Regularized Linear Model</div>
* Regularization is a way to reduce overfitting 
* Lower the degree of freedom harder it will be to overfit the data
* Way to regularize polynomial model is to reduce its degrees
* For linear model this is achieved by constraining weight of model
* this is done via 
> > * Ridge regression
> > * Lasso
> > * Elastic Net

### <div style="font-family: fantasy;">Ridge Regression</div>
* Also called $Tikonov$ $regularization$
* A regularizatin term $\alpha \sum_{i=1}^{n}\theta_i^2$ is added to the cost function only during the training.(train errors)
* Once training is done we need to use unregularized performance measures (val errors)
* good cost function should have optimization friendly derivatives while performance measure should be as close as possible to the final objective.
* $\alpha$ controls how much we want to regularize the CF.
* for $\alpha = 0$ it is linear regression
* for $\alpha$ very large all weights end up close to zero and result is a flat line going through data's mean 

$$
Ridge \ Regression \ cost \ function = 
J(\theta) = MSE(\theta) + \alpha \frac{1}{2} \sum_{i=1}^{n}\theta_i^2
$$
thus the regularization term is $\frac{1}{2}(||\boldsymbol w||_2)^2$ where $||\boldsymbol w||_2$ is $\ell_2$ norm of the weight vector. For gradient descent just add $\alpha \boldsymbol w$
* it is important to scale the data using ```StandardScaler``` before performing ridge regression
* increasing $\alpha$ increases bias
$$
ridge \ regression \ closed \ form \ solution = 
\hat{\boldsymbol\theta} = (\mathbf{X}^T \cdot \mathbf{X} + \alpha \cdot \boldsymbol A)^{-1} \cdot \mathbf{X}^T \cdot \boldsymbol y
$$
where $\boldsymbol A$ is n+1 $\times$ n+1 identitymatrix except with 0 in the top-cell corresponding to the bias term
here the equation uses matrix factorization technique by Adre Louis Chloesky

````python
np.random.seed(42)
m = 20
X = 3 * np.random.rand(m, 1)
y = 1 + 0.5 * X + np.random.randn(m, 1) / 1.5
X_new = np.linspace(0, 3, 100).reshape(100, 1)
plt.plot(X,y,'.')
````

````python
from sklearn.linear_model import Ridge
ridge_reg = Ridge(alpha=1,solver="cholesky")
ridge_reg.fit(X,y)
ridge_reg.predict([[1.5]]) # type: ignore
````

````python
# extra code ‚Äì this cell generates and saves Figure 4‚Äì17

def plot_model(model_class, polynomial, alphas, **model_kwargs):
    plt.plot(X, y, "b.", linewidth=3)
    for alpha, style in zip(alphas, ("b:", "g--", "r-")):
        if alpha > 0:
            model = model_class(alpha, **model_kwargs)
        else:
            model = LinearRegression()
        if polynomial:
            model = make_pipeline(
                PolynomialFeatures(degree=10, include_bias=False),
                StandardScaler(),
                model)
        model.fit(X, y)
        y_new_regul = model.predict(X_new)
        plt.plot(X_new, y_new_regul, style, linewidth=2,
                 label=fr"$\alpha = {alpha}$")
    plt.legend(loc="upper left")
    plt.xlabel("$x_1$")
    plt.axis([0, 3, 0, 3.5])
    plt.grid()

plt.figure(figsize=(9, 3.5))
plt.subplot(121)
plot_model(Ridge, polynomial=False, alphas=(0, 10, 100), random_state=42)
plt.ylabel("$y$  ", rotation=0)
plt.subplot(122)
plot_model(Ridge, polynomial=True, alphas=(0, 10**-5, 1), random_state=42)
plt.gca().axes.yaxis.set_ticklabels([])
plt.show()
````

### <div style="font-family: fantasy;"> Lasso Regression</div>
* least absolute shrinkage and selection operator regression
$$
Lasso \ Regression \ cost \ function = 
J(\theta) = MSE(\theta) + \alpha  \sum_{i=1}^{n} |\theta_i|
$$
* this uses $l1$ norm of the weight vector

````python
from sklearn.linear_model import Lasso

lasso_reg = Lasso(alpha=0.1)
lasso_reg.fit(X, y)
lasso_reg.predict([[1.5]]) # type: ignore
````

````python
plt.figure(figsize=(9, 3.5))
plt.subplot(121)
plot_model(Lasso, polynomial=False, alphas=(0, 0.1, 1), random_state=42)
plt.ylabel("$y$  ", rotation=0)
plt.subplot(122)
plot_model(Lasso, polynomial=True, alphas=(0, 1e-2, 1), random_state=42)
plt.gca().axes.yaxis.set_ticklabels([])
plt.show()
````

as we can see lasso regression completely eliminates the weights of the least important features\
even for a high degree polynomial(right figure) the graph with $\alpha = 0.01$ is almost linear but looks quadratic. This is because all the higher weight polynomial features are set to the weight zero or are completely eliminated

Lasso cost function is not differentiable at $\theta_i = 0$ (i.e. the bias condition) as $i=1,2,...n$ but gradient descent still works if we use a subgradient vector $\mathbf{g}$ instead when any $\theta_i = 0$
$$
    g(\theta,j) = \nabla_\theta MSE(\theta) + \alpha
    \begin{pmatrix}
        \operatorname{sign}(\theta_1) \\
        \operatorname{sign}(\theta_2) \\
        \operatorname{sign}(\theta_3) \\
                    \cdot \\
                    \cdot \\
                    \cdot \\
        \operatorname{sign}(\theta_n) \\ 
    \end{pmatrix}
    where 
    \operatorname{sign}(x) = \begin{cases}
        -1 & \text{if } x < 0 \\
        0 & \text{if } x = 0 \\
        1 & \text{if } x > 0 \\
    \end{cases}
$$

### <div style="font-family:fantasy;"> Elastic Net</div>
it is the middle ground of Ridge and Lasso. It's regularization term is the combination of both in some ratio $r$ called $mix \ ratio$. If $r=1$ it is pure Lasso and $r=0$ is pure Ridge. 

$$
Elastic \ Net \ cost \ function = 
J(\theta) = MSE(\theta) + r\alpha  \sum_{i=1}^{n} |\theta_i| + \frac{1-r}{2}\alpha \sum_{i=1}^{n}\theta_i^2
$$

````python
from sklearn.linear_model import ElasticNet
el_net = ElasticNet(alpha=0.1,l1_ratio=0.5)
el_net.fit(X,y)
el_net.predict([[1.5]]) # type: ignore
````

```l1_ratio``` hyperparamater is for r value corresponding to the $l1$ part i.e. Lasso

It is almost always preferable to use a little bit of reguarization. 
we use \
Ridge as a good default but if \
very few features are actually useful we should prefer Lasso or Elastic Net.\
They tend to reduce the weights of the useless features down to zero.\
Elastic Net is prefered over Lasso when number of features is greater than the number of training instances or wehn several features are strongly correlated.

### <div style="font-family: fantasy;r"> Early Stopping </div>
It is a different way to regularize iterative learning algos such as GD. It suggests to stop training as soon as the validation error reaches minimum. This is early stopping. 
below is an implementation for the same


````python
from sklearn.base import clone
sdg_reg = SGDRegressor(n_iter_no_change=1,warm_start=True,penalty=None,learning_rate="constant",eta0=0.0005)
minimum_val_error = float('inf')
X_train,X_val,y_train,y_val = train_test_split(X,y.ravel(),train_size=0.8)
X_train_poly = poly_features.fit_transform(X_train)
X_train_poly_scaled = std_scaler.fit_transform(X_train_poly.astype(np.float64))
X_val_poly = poly_features.fit_transform(X_val)
X_val_poly_scaled = std_scaler.fit_transform(X_val_poly.astype(np.float64))
best_epoch = None
best_model = None
for epoch in range(1000):
    sdg_reg.fit(X_train_poly_scaled,y_train)
    y_val_predict = sdg_reg.predict(X_val_poly_scaled)
    val_error = mean_squared_error(y_val_predict,y_val)
    if val_error < minimum_val_error:
        minimum_val_error = val_error
        best_epoch = epoch
        best_model = clone(sdg_reg)
````

# <div style="font-family: fantasy;"> Logistic Regression </div>
used to estimate the probability that an instance belongs to particular class. If more than 50% then positive class else negative class


### <div style="font-family: fantasy;"> Estimating Probabilities </div>
It computes the weighted sum of the input features(plus a bias term) but instead of outputting the result directly it outputs the _logistic_ of the result\
the probabily is computed as,
$$
\hat{p} = h_\theta(\boldsymbol X) = \sigma(\theta^T \cdot \boldsymbol X)
$$
where
$$
 \sigma(t) = \frac{1}{1+e^{-t}}
$$


````python
t = np.linspace(-10,10,100)
sig = 1/(1+np.exp(-t))
t,sig
````

````python
plt.title('sigmoid function')
plt.plot(t,sig,label=r"$\sigma(t) = \dfrac{1}{1 + e^{-t}}$")
plt.xlabel('t')
plt.ylabel(r'$\sigma$')
plt.plot([-10, 10], [0.5, 0.5], "k:")
plt.plot([-10,10], [1, 1], "k:")
plt.plot([-10,10], [0, 0], "k:")
plt.plot([0,0], [0, 1.0], "k:")
plt.axis([-10, 10, -0.1, 1.1])
plt.legend(loc="upper left")
````

now the prediction y can be defined as
$$
\operatorname{\hat{y}} = \begin{cases}
        0 & \text{if } \hat{p} < 0.5 \\
        1 & \text{if } \hat{p} >= 0.5 \\
\end{cases}
$$
note that $\sigma(t) < 0.5$ or y = 0 when t is negative \
$\sigma(t)>=0.5$ or y=1 when t is positive\
here t = $\theta^T \cdot \boldsymbol X$

### <div style="font-family: fantasy"> Training and cost function </div>
The objective of training is to set the parameter vector $\theta$ so that the model estimtes high probabilities for positive instances y=1 and low probabilities for negative instances y=0
$$
cost \ function \ of \ a \ single \ training \ instance
\operatorname{c}(\theta) = \begin{cases}
        -log(\hat{p}) & \text{if } y=1 \\
        -log(1-\hat{p}) & \text{if } y=0 \\
\end{cases}
$$
$-log(t)$ grows very large when t approaches 0 so the cost will be large if the model estimates a probability close to 0 for positive instance and it will also be very large if the model estimates probability close to 1 for negative instance (i.e. wrong estimation or errors). On the other hand, -log(t) is close to 0 when t is close 1 so the cost will be close to 0 if the estimated robability is close to 0 for a negative instance or close to 1 for a positive instance, which is precisely what we want.\
The cost function over the whole training set is simply the average cost overall training instances. It can be said as a log loss function
$$
J(\theta) = -\frac{1}{m} \sum_{i=1}^{m} [y^{(i)}log(\hat{p}^{(i)}) + (1-y^{(i)})log(1- \hat{p}^{(i)})]
$$

This cost function is covex thus GD is guaranted to find the global minimum. The partial derivatives of the cost function with regards to the $j^{th}$ model prameter $\theta_j$ is given by
$$
\frac{\partial}{\partial \theta_j} J(\theta) = \frac{1}{m} \sum_{i=1}^{m} (\sigma(\theta^T \cdot \boldsymbol X^{(i)}) - y^{(i)}) x_j^{(i)}
$$
you can use it generate the gradient vector once you get that you can use batch gradient descent.

### <div style="font-family: fantasy;"> Softmax Regression </div>
This is called multinomial Logistic Regression. It is used for classification with multiple classes.\
When given an instance $\boldsymbol x$ the softmax regression model first computes a score $s_k(\boldsymbol x)$ for each class k then estimates the probability of each class by applying softmax regression to the scores.
$$
s_k(\boldsymbol X) = \theta_k^T \cdot \boldsymbol X
$$
obviously each class does have its own set of parameters for classification $\theta_k$. All these vectors are stored as rows in the form of _parameter matrix_ $\mathbf{\Theta}$.\
Once the scores for every class are computed for the instance $\boldsymbol X$ we estimate the probability $\hat{p}_k$ that the instance belongs to the class k by running the scores through softmax function.
$$
\hat{p}_k = \sigma(s(\boldsymbol X))_k = \frac{e^{s_k(\boldsymbol X)}}{\sum_{j=1}^{K}e^{s_j(\boldsymbol X)}}
$$
* K is the number of classes 
* $s(\boldsymbol X)$ is a vector containing scores of each class for instance $\boldsymbol X$
* $\sigma(s(\boldsymbol X))_k$ is the estimated probability that instance $\boldsymbol x$ belongs to class k given the scores of each class for that instance
It predicts the class with highest probability i.e
$$
\hat{y} = argmax_k \sigma(s(\boldsymbol X))_k = argmax_k s_k(\boldsymbol X) = argmax_k(\theta_k^T \cdot \boldsymbol X)
$$
here the argmax function returns the value of k that maximizes the estimated probability $\sigma(s(\boldsymbol X))_k$

<div style="font-family: fantasy;font-size: medium;">training model</div>

the objective is to have a model that estimates high probabilities for the target class and low probability for the rest. Minimizing the cost function called _cross entropy_ leads to the objective. because it penalizes the model when it estimates low probability for target class. _cross entropy_ is frequently used to measure how well a set of estimated class proabilities match the target class.
$$
J(\Theta) = -\frac{1}{m} \sum_{i=1}^{m} \sum_{k=1}{K} y_k^{(i)} log(\hat{p}_k^{(i)})
$$
$y_k^{(i)}$ is 1 if the target class for the $i^{th}$ is k else it is 0
now the gradient vector of the cost fuction wrt $\theta_k$ is 
$$
\nabla_{\theta_k}J(\Theta) = \frac{1}{m} \sum_{i=1}^{m}(\hat{p}_k^{(i)} - y_k^{(i)})\mathbf{x}^{(i)}
$$
now we can compute the gradient vector for every class k in range 1 to K and use it in GD to find parameter matrix $\Theta$ that minimizes the cost function\


````python
X = iris["data"][:,(2,3)] # type: ignore
y = iris["target"] # type: ignore
softmax_reg = LogisticRegression(multi_class="multinomial",solver="lbfgs",C=10)
softmax_reg.fit(X,y)
````
